# <font color = blue>Part A - Classification

## 1. Loading Data and Manipulating with Pandas

Instructions:

- Load iris dataset from sklearn.datasets module 




```python
#Your solution here
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
idf = pd.DataFrame(iris.data, columns=iris.feature_names)
```

## 2. Train test split

Instructions:

- Split your datasets into train and test set




```python
#Your solution here
from sklearn.model_selection import train_test_split
x = idf
y = iris.target
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5, stratify=y)
```

## 3. Model Training

Instructions:

- Import DecisionTreeClassifier Algorithm from sklearn.





```python
#Your solution here
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=3, criterion='entropy')

#train the model
clf = dtc.fit(x_train,y_train)
clf
```

## 3. Model Evaluation

Instructions:

- Use test set to evaluate your model's performance.





```python
#Your   here
dtc.score(x_train, y_train)
```

```python
dtc.score(x_test, y_test)
```

## 4. Plot Tree

Instructions:

- Use sklearn tree module to plot the tree.





```python
# write your solution here
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plot_tree(clf)
plt.show()
```

## 5. Make Prediction

Instructions:

- Use sample data to make prediction





```python
#write your solution here
x_sample = [[5.2,4.2,1.4,0.5]]
val = iris.target_names[dtc.predict(x_sample)]
val[0]
```

# <font color = blue>Part B - Regression

## 1. Loading Data and Manipulating with Pandas

Instructions:

- Load boston dataset from sklearn.datasets module 




```python
#Your solution here
from sklearn.datasets import load_boston
boston = load_boston()
bdf = pd.DataFrame(boston.data, columns = boston.feature_names)
```

## 2. Train test split

Instructions:

- Split your datasets into train and test set




```python
#Your solution here
from sklearn.model_selection import train_test_split
X = bdf[['RM', 'LSTAT','B','PTRATIO']]
Y = boston.target
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=5)
```

## 3. Model Training

Instructions:

- Import DecisionTreeRegressor Algorithm from sklearn.




```python
#Your solution here
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth=4,criterion='mse')
dtr = dtr.fit(X_train, Y_train)
```

## 3. Model Evaluation

Instructions:

- Use test set to evaluate your model's performance.





```python
#Your solution here
dtr.score(X_train, Y_train)
```

```python
dtr.score(X_test, Y_test)
```

## 4. Plot Tree

Instructions:

- Use sklearn tree module to plot the tree.





```python
#write your solution here
plt.figure(figsize=(10,10), dpi=120)
plot_tree(dtr)
plt.show()
```

## 5. Make Prediction

Instructions:

- Use sample data to make prediction





```python
#write your solution here
x_sam = [[10.34,5.89,400,21]]
dtr.predict(x_sam)
```
