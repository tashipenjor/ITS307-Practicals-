## 4. Python Data Structures

4.1.Define a function `wordCount()` which will accepts a sentence as parameter
from user and count the occurrence of each words in the given sentence and return dictionary with word as `key` and count of word as `value`.

“This is python programming. It is very good programming language”



```python
#write your solution here
def wordCount(sentence):
    words = sentence.split()
    words = [word.replace(".","") for word in words]
    result = {}
    
    for word in words:
        if word in result:
            result[word] += 1
        else:
            result[word] = 1
        
    return result

print(wordCount("This is python programming. It is very good programming langauge."))

```

4.2. Write a function called courseInfo() that takes as input a list of
tuples. Each tuple contains two items: the first item in each tuple is a
student’s name as a string, and the second item in each tuple is that
student’s age as an integer.
The function should return a dictionary with two keys. The key “students”
should have its value a list of all the students. The key “ave_age” should
have as its value a float representing the average age of all the students
in the list.



```python
#write your code here
def courseInfo(studentInfo):
    names = []
    total_age = 0
    for student in studentInfo:
        names.append(student[0])
        total_age += student[1]
    result = {"students":names,"avg_age":total_age/len(studentInfo)}
    return result

#Donot edit code here
assert courseInfo([("Jackie", 20), ("Marguerite", 21)]) == {"students": ['Jackie', 'Marguerite'], "avg_age": 20.5}
assert courseInfo([("Nima", 25), ("Pema", 24),("Namgay",23)]) == {"students":['Nima','Pema','Namgay'],"avg_age":24}
assert courseInfo([('Sagar',18), ("Sonam",19), ("Tashi",20),("Dechen",25), ("Kelden",33)]) == {'students': ['Sagar', 'Sonam', 'Tashi', 'Dechen', 'Kelden'], 'avg_age': 23.0}
print("All the testcases passed!")
```

4.3. Write a function calculateScore() which takes two parameters: a
dictionary of student answers and a dictionary of correct answers. Both
dictionaries should have integers as their keys, and one character string as
their values.
calculate_score() should count how many questions the student got
right. Or how many keys for which the associated value is the same in the
student’s dictionary and in the answer key dictionary.



```python
#write your solution here
def calculateScore(std_answer, correct_answer):
    count = 0
    for answer in std_answer:
        if(answer in std_answer and std_answer[answer] == correct_answer[answer]):
            count += 1
    return count


#Donot edit code section here
correct_Answer = {2: "A", 1: "A", 3: "A", 4: "D", 5: "B",6:'B',7:'C',8:'D',9:'C',10:'A'}
student1_Answer = {1: "A", 2: "B", 3: "C", 4: "D", 5: "A",6:'B',7:'C',8:'D', 9:'C',10:'D'}
student2_Answer = {1: "A", 2: "B", 3: "C", 4: "D", 5: "E",6:'A',7:'B',8:'C',9:'D',10:'B'}
student3_Answer = {1: "C", 2: "B", 3: "C", 4: "A", 5: "E",6:'C',7:'D',8:'B',9:'D',10:'D'}

assert calculateScore(student1_Answer, correct_Answer) == 6
assert calculateScore(student2_Answer, correct_Answer) == 2
assert calculateScore(student3_Answer, correct_Answer) == 0
print("All the testcases passed!")
    
```

Develop a student management system for GCIT. You are required to store students’ details such as name, age, and marks. Make sure that your system has all CRUD operations implemented.

- Add Student Details
- Delete Student Details
- View Student Details
- Update student Details



```python
import pandas as pd 
Student_GCIT = {'Name':['Tashi','Namgay','Pema','Thinley','Tshering'],
                                  'Age':[23,25,23,28,20],'Marks':[65,53,75,45,80]}
def addStdDetails(name,age,marks):
    Student_GCIT['Name'].append(name)
    Student_GCIT['Age'].append(age)
    Student_GCIT['Marks'].append(marks)
    return pd.DataFrame(Student_GCIT)
    
def deleteStdDetails(name):
    index = Student_GCIT['Name'].index(name)
    index = int(index)
    del Student_GCIT['Name'][index]
    del Student_GCIT['Age'][index]
    del Student_GCIT['Age'][index] 
    return Student_GCIT
    
    
def updateStdDetails(name, new_name, new_age, new_marks):
    index = Student_GCIT['Name'].index(name)
    index = int(index)
    Student_GCIT['Name'][index] = new_name
    Student_GCIT['Age'][index] = new_age
    Student_GCIT['Marks'][index] = new_marks
    return pd.DataFrame(Student_GCIT)

print('A student management system for GCIT:\n',pd.DataFrame(Student_GCIT),'\n')
print("Commands are::\n1: Add student details\n2: Update the student details\n3: Delete the student details\n0: To Exit")  

while True:
    userInp = int(input("Enter command: "))
    if(userInp == 1):
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        marks = int(input("Enter marks: "))
        print('New stdudent details are added!\n',addStdDetails(name, age, marks))
    elif(userInp == 2):
        name = input("Enter the student name to Update: ")
        new_name = input("Enter your new name: ")
        new_age = input("Enter your new age: ")
        new_marks = input("Enter your new marks: ")
        print('Stdudent details are updated!\n',updateStdDetails(name, new_name, new_age, new_marks))
    elif(userInp == 3):
        name = input("Enter the student name to Delete: ")
        print('Student details are deleted!',deleteStdDetails(name))
        
    elif(userInp == 0):
        break
```
